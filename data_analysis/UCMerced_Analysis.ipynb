# UCMerced Dataset Analysis

### Imports

```python
import pandas as pd
from torchvision import datasets, transforms, utils
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import torch
```

### Read In Dataset and Convert to Pandas DataFrame

```python
dataset = datasets.ImageFolder(root="../data/UCMerced_Landuse_Images")

data_list = []

pixels_list = []

for path, label in dataset.samples:
    img = Image.open(path)
    img_pixels = list(img.getdata())

    img_array = np.array(img)
    
    # Calculate statistics on pixel values
    mean_value = np.mean(img_array)
    std_dev = np.std(img_array)
    min_value = np.min(img_array)
    max_value = np.max(img_array)
    
    data_list.append({
        "image_path": path,
        "label": label,
        "mean_value": mean_value,
        "std_dev": std_dev,
        "min_value": min_value,
        "max_value": max_value,
    })

    pixels_list.append(img_pixels)
    
    # data_list.append({"image_path": path, "label": label, "pixels": img_pixels})

df = pd.DataFrame(data_list)

pixels_df = pd.DataFrame(pixels_list)

print(df.head())
print(pixels_df.head())
```

```python
dataset = datasets.ImageFolder(root="../data/UCMerced_Landuse_Images", transform=transforms.ToTensor())

# Create a DataLoader to iterate through the dataset
dataloader = torch.utils.data.DataLoader(dataset, batch_size=1, shuffle=True)

# Function to show images in a grid with class labels
def imshow_grid(images, labels, class_names, title):
    fig, ax = plt.subplots(figsize=(10, 10))
    images = utils.make_grid(images, nrow=3)
    npimg = images.numpy()
    ax.imshow(np.transpose(npimg, (1, 2, 0)))
    
    # Add class labels
    for i in range(3):
        for j in range(3):
            ax.text((j+0.05) * images.size(2) / 3, (i+0.86) * images.size(1) / 3, class_names[labels[i * 3 + j]],
                     ha='left', va='top', color='white', fontsize=8, bbox=dict(facecolor='black', alpha=0.8))
    
    ax.set_title(title)
    plt.show()

# Plot nine images in a 3 by 3 grid with class labels
images_to_plot = []
labels_to_plot = []

for i, (images, labels) in enumerate(dataloader):
    images_to_plot.append(images)
    labels_to_plot.append(labels.item())

    if i == 8:  # Display 9 images
        break

class_names = dataset.classes
imshow_grid(torch.cat(images_to_plot, dim=0), labels_to_plot, class_names, title="UCMerced Landuse Images with Class Labels")
```
